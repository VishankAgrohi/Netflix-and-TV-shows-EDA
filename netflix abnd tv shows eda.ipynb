## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import pandas as pd

# Load the dataset
df = pd.read_csv("NETFLIX MOVIES AND TV SHOWS CLUSTERING.csv")

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Handle missing values
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Unknown", inplace=True)
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna("Not Rated", inplace=True)

# Extract duration in numeric format
df['duration_type'] = df['duration'].str.extract(r'([a-zA-Z]+)')
df['duration_int'] = df['duration'].str.extract(r'(\d+)').astype(float)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Optional: Feature engineering - extract year added
df['year_added'] = df['date_added'].dt.year

# Final clean data preview
print(df.info())
print(df.head())
````

### What all manipulations have you done and insights you found?

Answer ✅ Data Wrangling Steps Performed:
Standardized Column Names:

Converted column names to lowercase and replaced spaces with underscores for consistency (e.g., show_id instead of Show ID).

Converted Date Format:

Converted the date_added column from string to datetime format to enable time-based analysis.

Handled Missing Values:

Replaced missing values in:

director, cast, and country with "Unknown".

rating with "Not Rated" (to preserve rows for analysis).

Left date_added as NaT where parsing failed, for potential filtering.

Extracted Duration Components:

Split the duration column into:

duration_int: numeric part (e.g., 90)

duration_type: string part (min or Season(s)) to distinguish between movies and TV shows.

Removed Duplicates:

Dropped any exact duplicate rows to ensure data quality.

Feature Engineering:

Created a year_added column from date_added to analyze content trends over time.

🔍 Initial Insights:
Type Distribution:

The dataset includes both Movies and TV Shows, with movies likely dominating in volume.

Duration Trends:

TV shows use "Seasons" while movies list duration in "minutes" — useful for content type separation.

Missing Data Insight:

A significant number of entries are missing director and cast information, possibly due to regional content or older releases.

Country Insights:

United States, India, and a few others frequently appear in the country column — suggests content dominance by these regions.

Release vs. Added to Netflix:

Comparing release_year to date_added can help analyze how long it takes for content to appear on Netflix after release.

Ratings Overview:

Presence of Not Rated, TV-MA, PG-13, etc., allows categorization into adult vs family-friendly content.
